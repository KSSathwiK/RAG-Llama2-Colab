```python
import torch
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core import PromptTemplate

# Model names (make sure you have access on HF)
LLAMA2_7B = "meta-llama/Llama-2-7b-hf"
LLAMA2_7B_CHAT = "meta-llama/Llama-2-7b-chat-hf"
LLAMA2_13B = "meta-llama/Llama-2-13b-hf"
LLAMA2_13B_CHAT = "meta-llama/Llama-2-13b-chat-hf"
LLAMA2_70B = "meta-llama/Llama-2-70b-hf"
LLAMA2_70B_CHAT = "meta-llama/Llama-2-70b-chat-hf"

selected_model = LLAMA2_7B_CHAT

SYSTEM_PROMPT = """You are an AI assistant that answers questions in a friendly manner, based on the given source documents. Here are some rules you always follow:
- Generate human readable output, avoid creating output with gibberish text.
- Generate only the requested output, don't include any other language before or after the requested output.
- Never say thank you, that you are happy to help, that you are an AI agent, etc. Just answer directly.
- Generate professional language typically used in business documents in North America.
- Never generate offensive or foul language.
"""

query_wrapper_prompt = PromptTemplate(
    "[INST]<>\n" + SYSTEM_PROMPT + "<>\n\n{query_str}[/INST] "
)

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=2048,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name=selected_model,
    model_name=selected_model,
    device_map="auto",
    # change these settings below depending on your GPU
    model_kwargs={"torch_dtype": torch.float16, "load_in_8bit": True},
)
```

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")
```

```python
from llama_index.core import Settings

Settings.llm = llm
Settings.embed_model = embed_model
```

```python
# Function to calculate tax under the old regime
def calculate_tax_old_regime(income):
    if income <= 250000:
        return 0  # No tax for income below 2.5 lakhs
    elif 250000 < income <= 500000:
        return (income - 250000) * 0.05  # 5% of excess over 2.5 lakhs
    elif 500000 < income <= 1000000:
        return 12500 + (income - 500000) * 0.2  # 12,500 + 20% of excess over 5 lakhs
    else:
        return 112500 + (income - 1000000) * 0.3  # 1,12,500 + 30% of excess over 10 lakhs


```

```python
# Function to calculate tax under the new regime
def calculate_tax_new_regime(income):
    if income <= 300000:
        return 0  # No tax for income below 3 lakhs
    elif 300000 < income <= 600000:
        return (income - 300000) * 0.05  # 5% of excess over 3 lakhs
    elif 600000 < income <= 900000:
        return 15000 + (income - 600000) * 0.1  # 15,000 + 10% of excess over 6 lakhs
    elif 900000 < income <= 1200000:
        return 45000 + (income - 900000) * 0.15  # 45,000 + 15% of excess over 9 lakhs
    elif 1200000 < income <= 1500000:
        return 90000 + (income - 1200000) * 0.2  # 90,000 + 20% of excess over 12 lakhs
    else:
        return 150000 + (income - 1500000) * 0.3  # 1,50,000 + 30% of excess over 15 lakhs
```

```python
def ask_for_tax_calculation():
    try:
        income = int(input("Please enter your income in digits: "))
        regime = input("Please specify tax regime (old/new) or press both: ").lower()

        if regime == "old":
            tax_old = calculate_tax_old_regime(income)
            print(f"Your calculated tax under the old regime is: ₹{tax_old}")
        elif regime == "new":
            tax_new = calculate_tax_new_regime(income)
            print(f"Your calculated tax under the new regime is: ₹{tax_new}")
        else:
            # Show both calculations if regime is not specified
            tax_old = calculate_tax_old_regime(income)
            tax_new = calculate_tax_new_regime(income)
            print(f"Your calculated tax under the old regime is: ₹{tax_old}")
            print(f"Your calculated tax under the new regime is: ₹{tax_new}")

    except ValueError:
        print("Please enter a valid number for income.")
```

```python
# Function to handle user queries and identify tax-related queries
import re

def handle_user_query(query):
    query = query.lower()

    # Define a list of keywords related to tax calculation
    tax_keywords = ["calculate tax", "what is my tax", "tax deduction", "how much tax", "income tax", "tax amount"]

    # Check for tax-related keywords
    if any(keyword in query for keyword in tax_keywords):
        ask_for_tax_calculation()

    # Use regex to match different possible queries
    elif re.search(r"\btax\b.*\b(calculate|deduction|how much|amount|my)\b", query):
        ask_for_tax_calculation()

    else:
        # Use query engine for other queries
        response = query_engine.query(query)
        print(response)
```
